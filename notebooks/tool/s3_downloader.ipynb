```python
# If gcsfs or zarr is missing, install them (this line is for Jupyter notebooks)
%pip install gcsfs zarr

import xarray as xr
import gcsfs

# Use gcsfs to get a mapper for the GCS Zarr store and open with xarray
fs = gcsfs.GCSFileSystem()
mapper = fs.get_mapper('weatherbench2/datasets/ifs_extended_range/daily/ifs-ext-full-single-level-avg_ens_mean.zarr')

# Open the zarr store. Use consolidated=True if the store has a consolidated metadata file.
ds = xr.open_zarr(mapper, consolidated=True)
ds
```
